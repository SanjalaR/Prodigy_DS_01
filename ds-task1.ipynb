```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv("/kaggle/input/indiancities/Indian_cities.csv")
df
```

```python
gp1 = df.groupby('state_name').sum()
gp1
```

```python
pop = gp1[['population_male', 'population_female', 'population_total']].sort_values('population_total', ascending=False).drop('population_total', axis=1)
pop
```

```python
ax = pop.plot(kind='bar', figsize=(15,8))
ax.set_xlabel('STATE')
ax.set_ylabel("POPULATION in 10^7")
```
